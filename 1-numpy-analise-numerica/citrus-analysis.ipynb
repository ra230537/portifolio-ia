```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('citrus.csv', delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
orange_diameter = data[:5000, 0]
orange_weight = data[:5000, 1]
grapefruit_diameter = data[5000:, 0]
grapefruit_weight = data[5000:, 1]
```

```python
plt.plot(orange_diameter, orange_weight)
plt.plot(grapefruit_diameter, grapefruit_weight)
plt.legend(['Orange', 'grapefruit'])
```

Find y = a * x + b to orange and grapefruit

```python
def find_angular_coeficient(x, y):
    y_mean = np.mean(y)
    x_mean = np.mean(x)
    return np.sum((x - x_mean) * (y - y_mean)) / np.sum(np.power(x - x_mean, 2))
def find_linear_coeficient(x, y, a):
    y_mean = np.mean(y)
    x_mean = np.mean(x)
    return y_mean - a * x_mean
orange_a = find_angular_coeficient(orange_diameter, orange_weight)
orange_b = find_linear_coeficient(orange_diameter, orange_weight, orange_a)
orange_y = orange_a * orange_diameter + orange_b

grapefruit_a = find_angular_coeficient(grapefruit_diameter, grapefruit_weight)
grapefruit_b = find_linear_coeficient(grapefruit_diameter, grapefruit_weight, grapefruit_a)
grapefruit_y = grapefruit_a * grapefruit_diameter + grapefruit_b
```

```python
# Lets plot this
plt.plot(orange_diameter, orange_weight)
plt.plot(orange_diameter, orange_y)
plt.legend(['Original Value', 'Linear Regression'])
np.linalg.norm(orange_weight - orange_y)
```

```python
plt.plot(grapefruit_diameter, grapefruit_weight)
plt.plot(grapefruit_diameter, grapefruit_y)
plt.legend(['Original Value', 'Linear Regression'])
np.linalg.norm(grapefruit_weight - grapefruit_y)
```
