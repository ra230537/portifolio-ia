```python
# lidando com dados nulos
df.isnull() # Booleano mostrando quais dados são e não são nulos
```

```python
# Para ver quais dados nulos, basta usar a função sum
df.isnull().sum()
```

```python
# Vamos substituir os valores nulos pelo valor 0 nesse caso
df = df.fillna(0)
df
```

```python
# conferindo
df.isnull().sum()
```

Outras abordagens para remover dados nulos são:
1. Remover com dropna()
2. preencher com algum dado especifico com o fillna, da pra usar o metodo bfill ou ffill dentro do parametro para usar o valor anterior ou posterior na hora de preencher
3. Da pra interpolar os valores com a função interpolate

```python
# Remover registros inconsistentes, valor igual a 0 ou condominio 
df.query('Valor == 0 | Condominio == 0')
```

```python
# Remover registros inconsistentes, valor igual a 0 ou condominio 
registros_remover = df.query('Valor == 0 | Condominio == 0').index
registros_remover
```

```python
df.drop(registros_remover, axis=0, inplace=True)
```

```python
# Mostrou que tudo foi removido
df.query('Valor == 0 | Condominio == 0').index
```

```python
# Vamos remover a coluna do apartamento porque so temos apartamento
df.drop('Tipo', axis=1, inplace=True)
df
```

Aplicando filtros

```python
# Apartamentos que possuem 1 quarto e alguel menor que 1200
df['Quartos'] == 1
```

```python
# Podemos passar uma series dentro do dataframe
df[df['Quartos'] == 1 ]
```

```python
df[df['Valor'] < 1200 ]
```

```python
# Juntando
df_1 = df[(df['Valor'] < 1200) & (df['Quartos'] == 1)]
df_1
```

```python
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df_2 = df[selecao]
df_2
```

# Parte 4: Manipulando dados

```python
df = pd.read_csv('rent.csv', sep=';')
df = df.fillna(0)
```

```python

df['Valor_mes'] = df['Valor'] + df['Condominio']
df
```

```python
df['Valor_ano'] = df['Valor_mes'] * 12 + df['IPTU']
df
```

```python
df['Descricao'] = df['Tipo'] + ' em ' + df['Bairro'] + ' com ' + \
                                        df['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + df['Vagas'].astype(str) + ' vaga(s) de garagem.'

df
```

```python
df['Possui_suite'] = df['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
df
```
