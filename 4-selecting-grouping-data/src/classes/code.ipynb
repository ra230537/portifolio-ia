# Agrupamento de dados

```python
# select colunas ate  produto
cols_info = list(df.loc[:, 'Nível 1 - Setor': 'Produto'].columns)
cols_info
```

```python
gas_emissoes = list(df.loc[:, 1970:2021].columns)
gas_emissoes
```

```python
# fazendo a transformação dessas varias colunas em 2: header & value
df_year = df.melt(id_vars= cols_info, value_vars=gas_emissoes, var_name='Ano', value_name='Emissao')
df_year
```

Wide = tabela pivo, tem várias colunas com vários atributos, como eh o caso de df
Long = formato normalizado, geralmente tem duas colunas principais de informação: uma parao nome das variaveis e outra para o valor das variaáveis, como foi o caso acima de df_year

## Obter os gases mais emitidos

```python
df_year.groupby('Gás').groups
```

```python
# get specific group
df_year.groupby('Gás').get_group('CO2 (t)')
```

```python
# get specific group
emissao_gas = df_year.groupby('Gás')[['Emissao']].sum().sort_values('Emissao', ascending = False)
emissao_gas
```

```python
emissao_gas.plot(kind='barh', figsize=(10, 6));
```

```python
#get values of co2
gas_co2 = emissao_gas.iloc[0:9].sum()
total = emissao_gas.sum()
print(f'Emissao de CO2 corresponde a {float((gas_co2/total).iloc[0])*100:.2f}% de emissão total de gases estufa no Brasil de 1970 a 2021 ')
```
