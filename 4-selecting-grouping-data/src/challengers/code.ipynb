```python
# get text columns
text_cols = list(df.loc[:, 'Nível 1 - Setor':'Produto'].columns)
value_cols = list(df.loc[:,1970: 2021].columns)
df_long = df.melt(id_vars=text_cols, value_vars=value_cols, var_name='Ano',value_name='Emissao')
df_long
```

```python
df_long[df_long['Ano'] == 2021].groupby('Nível 1 - Setor')[['Emissao']].mean()
```

```python
df_long.groupby('Nível 1 - Setor')[['Emissao']].sum().sort_values(by='Emissao',ascending=False)
```
